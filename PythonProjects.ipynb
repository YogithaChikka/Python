```python
questions = [ 
    {
        "prompt" : "What is the capital of France?", 
        "options" : ["A. Paris", "B. London", "C. Berlin", "D. Madrid"],
        "answer" : "A"
    },

    {   
        "prompt" : "Which language is primarily spoken in Brazil?",
        "options" : ["A. Portuguese", "B. French", "C. Chinese", "D. Spanish"],
        "answer" : "A"
    },

    {
        "prompt" : "What is the smallest prime number?",
        "options" :["A. 1", "B. 2", "C. 3", "D. 5"],
        "answer" : "B"
    },

    {
        "prompt" : "Who wrote 'To Kill a Mockingbird'?",
        "options" : ["A. Harper Lee", "B. Mark Twain", "C. J.K. Rowling", "D. Ernest Hemmingway"],
        "answer" : "A"
    }
]

def run_quiz(questions):
    score = 0
    for question in questions:
        print(question["prompt"])
        for option in question["options"]:
            print(option)
        answer = input("Enter your answer (A, B, C or D): ").upper()
        while answer not in ("A", "B", "C","D"):
            print("Enter valid option. The option should be either A, B, C or D \n")
            answer = input("Enter your answer (A, B, C or D): ").upper()
       
        if answer == question["answer"]:
            print("Correct!!\n")
            score += 1
        else:
            print("Wrong :( The correct answer is", question["answer"], "\n")
        
            
    if score == len(questions):
        print(f"Brillianttt!! You got {score} out of {len(questions)} questions correct.")
    else:
        print(f"Stop the guess game, malli chadukoni raa ra chitti\nYou got {score} out of {len(questions)} questions correct.")

print(" ")
print("HEY LET'S PLAY QUIZ.\n\n")
run_quiz(questions)
```
